## t-test

```python
X = df_2.drop('priceUSD', axis = 1)
y = df_2['priceUSD']
X_train, X_test, y_pop, y_sample = train_test_split(X, y, test_size=0.005, random_state=42)
```

```python
print('The mean price of cars is:', y.mean())
```

```python
print('The mean price and standard deviation of sample size is:', y_sample.mean(), 'and', y_sample.std(), 'respectively')
```

We have taken 0.5% of total population randomly as sample for t-test, mean and standard deviation is known for sample which are 8278.965 and 11400.67 repectively.  
Let's say we do not know the standard deviation of cars but mean is known, which is 7430.38.  
We wish to perform t-test for 5% significance level.

### Hypothesis Formulation
Ho: mu equals muo  
Ha: mu not equals muo

```python
## Variables:
mu = 7430.38      # Mean of population
alpha = 0.05      # Significnce level
x_bar = 8278.965  # Sample mean
n = 200           # Number of samples
s = 11400.67      # Sample stanard deviation

#t = test static
t = (x_bar - mu)/(s/np.sqrt(n))  
print('Value of test static is:', t)
```

```python
p_value = st.t.cdf(1.05, 199)  # p_value @ test static t, 199 is the degrees of freedom (n-1).
print('p_value at test static t is:', p_value)
```

##### As p_value(0.85) > alpha(0.05), this means that we do not have enough evidence to reject null hypothesis, so we have to say that do not Ho.   
##### We have to say mu = muo  (mu = 7430.38).
